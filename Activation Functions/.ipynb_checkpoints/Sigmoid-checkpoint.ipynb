```python
import numpy as np
import matplotlib.pyplot as plot

# Sigmoid function
def sigmoid(X):
    return 1 / (1+np.exp(-X))

# Create values for X
X = np.arange(-10, 10, 0.1)

# Plot graph using X array and Y as the sigmoid
plot.plot(X, sigmoid(X))

plot.title('Sigmoid Function')
plot.xlabel('Values')
plot.ylabel('Sigmoid(Values)')

plot.show()
```
